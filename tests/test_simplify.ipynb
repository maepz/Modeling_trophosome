## To simplify the population network

```python

from project_package.simplify import remove_empty_leaves_and_rescale_edges
########

fitnesses=[1,0.9] # allele_id:fitness
abundances=[30,70] # allele_id:abundance
n=2

G = nx.complete_bipartite_graph(1,n) # create empty graph of n strains

print(G.nodes)
for i in range(len(G.nodes)): # for each strain, add fitness and abundance data
    G.add_node(i, abundance=abundances[i-1],fitness=fitnesses[i-1])
for start,end in G.edges:
    G.add_edge(start,end,distance=999)
G.update(edges=[[2,n+1,{'distance':1}],[2,n+2,{'distance':1}],[n+2,n+3,{'distance':1}],[n+3,n+4,{'distance':1}],[n+4,n+5,{'distance':1}]], nodes=[[n+1,{'abundance':0,'fitness':1}],[n+2,{'abundance':0,'fitness':1}],[n+3,{'abundance':0,'fitness':1}],[n+4,{'abundance':0,'fitness':1}],[n+5,{'abundance':10,'fitness':1}]])

visualize_pop(G,show_empty=True)

newG=remove_empty_leaves_and_rescale_edges(G)
plt.figure()
visualize_pop(newG,show_empty=True)
```
