# Description

This script is for a simple Fisher-Wright model of evolution that includes fitness. The whole population is replaced at each t+1

## Without fitness; The whole population is replaced at each t+1

```python

########################

def update_pop(population):
    from random import choices
    from collections import Counter

    alleles = list(population.keys())
    weights=list(population.values()) # The sampling of alleles into the next generation is weighted by their relative frequencies
    n=sum(weights)

    new_gen = choices(alleles, weights, k=n) 

    new_pop=Counter(new_gen)

    return(new_pop)

def run_until_fixation(population):
    results={0:population}

    t=0
    while True:
        t+=1
        population_tplus1=update_pop(population)
        results.update({t:population_tplus1})

        population=population_tplus1

        if len(results[t])==1:
            return(results)
            break
        

#####################

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


population={1:30,2:70} # allele:abundance
n=sum(list(population.values()))


for trial in range(10):
    
    results=run_until_fixation(population)
    df=pd.DataFrame.from_dict(results, orient='index').melt(ignore_index = False)
    
    my_palette={'1':'blue','2':'red'}
    ax=sns.lineplot(data=df,x=df.index,y=df.value/n,hue=map(str,df.variable), palette=my_palette)
    ax.set_ylim(0, 1)
    ax.set(xlabel='Generations', ylabel='Relative abundance of alleles')
    ax.get_legend().remove()
```

## With fitness; the whole population is replaced at t+1

```python

########################

def update_pop(population,fitness_dic):
    from random import choices
    from collections import Counter

    alleles = list(population.keys())
    abs_weights=[population[x]*fitness_dic[x] for x in population.keys()] # absolute fitness
    rel_weights=[wt/sum(abs_weights) for wt in abs_weights] # relative fitness
   
    weights=rel_weights # using relative fitness instead of absolute fitness ensures the population size stays the same

    n=sum(population.values())
    
    new_gen = choices(alleles, weights, k=int(n))

    new_pop=Counter(new_gen)

    return(new_pop)

def run_until_fixation(population,fitness_dic):
    results={0:population}

    t=0
    while True:
        t+=1
        population_tplus1=update_pop(population,fitness_dic)
        results.update({t:population_tplus1})

        population=population_tplus1

        if len(results[t])==1:
            return(results)
            break
        

#####################

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline

fitness_dic={1:1,2:0.9}
population={1:30,2:70} # allele:abundance
n=sum(list(population.values()))


for trial in range(10):
    
    results=run_until_fixation(population,fitness_dic)
    df=pd.DataFrame.from_dict(results, orient='index').melt(ignore_index = False)
    
    my_palette={'1':'blue','2':'red'}
    ax=sns.lineplot(data=df,x=df.index,y=df.value/n,hue=map(str,df.variable), palette=my_palette)
    ax.set_ylim(0, 1)
    ax.set(xlabel='Generations', ylabel='Relative abundance of alleles')
    ax.get_legend().remove()
```
